```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.bbc.com/sport/olympics/paris-2024/medals'
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
webpage=requests.get(url,headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'html')
```

```python
countries=soup.find_all('span',class_='ssrcss-ymac56-CountryName ew4ldjd0')
```

```python
countries_list=[name.text for name in countries]
```

```python
countries_list
```

```python
rank=soup.find_all('td',class_='ssrcss-fvkmzs-StyledTableData ef9ipf1')
```

```python
rank_list=[ranks.text for ranks in rank[0::6]]
```

```python
rank_list
```

```python
data_list=soup.find_all('td',class_='ssrcss-fvkmzs-StyledTableData ef9ipf1') #from here we can extract medal counts respectively
```

```python
gold_list=[golds.text for golds in data_list[2::6]]
```

```python
gold_list
```

```python
silver_list=[silvers.text for silvers in data_list[3::6]]
```

```python
silver_list
```

```python
bronze_list=[bronze.text for bronze in data_list[4::6]]
```

```python
bronze_list
```

```python
total_medals_list=[total.text for total in data_list[5::6]]
```

```python
total_medals_list
```

```python
df=pd.DataFrame({'Rank':rank_list,
                 'Country':countries_list,
                 'Gold':gold_list,
                 'Silver':silver_list,
                'Bronze':bronze_list,
                 'Total Medals':total_medals_list
                })
```
